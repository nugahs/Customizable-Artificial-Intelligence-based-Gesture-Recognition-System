```python
import cv2
import time
import numpy as np
import mediapipe as mp
import matplotlib.pyplot as plt
import openpyxl as ex
from selenium import webdriver                                                    #Required to use delay functions
import pyautogui
from selenium.webdriver.chrome.options import Options
chrome_options = Options() 
chrome_options.add_experimental_option("detach", True)

#code for category 1
def category_one(i):
    driver = webdriver.Chrome(options=chrome_options)
    driver.maximize_window()
    driver.get(Sheet.cell(row=i, column=3).value)
    
    
#code for category 2
def category_two(i):
    driver = webdriver.Chrome(options=chrome_options)
    driver.maximize_window()
    driver.get(Sheet.cell(row=i, column=3).value)                              #website 
    email =driver.find_element_by_id(Sheet.cell(row=i, column=4).value)        #email field
    email.send_keys(Sheet.cell(row=i, column=5).value)                         #email
    password =driver.find_element_by_id(Sheet.cell(row=i, column=7).value)     #password field
    password.send_keys(Sheet.cell(row=i, column=8).value)                      #password
    loginbutton =driver.find_element_by_name(Sheet.cell(row=i, column=9).value)  #loginbutton
    loginbutton.click()                                                        #click login button    

    
#code for category 3
def category_three(i):
    driver = webdriver.Chrome(options=chrome_options)
    driver.maximize_window()
    driver.get(Sheet.cell(row=i, column=3).value)                              #website 
    email =driver.find_element_by_id(Sheet.cell(row=i, column=4).value)        #email field
    email.send_keys(Sheet.cell(row=i, column=5).value)                         #email
    nextbutton =driver.find_element_by_id(Sheet.cell(row=i, column=6).value)   #next button
    nextbutton.click()                                                         #click next button
    password =driver.find_element_by_id(Sheet.cell(row=i, column=7).value)     #password field
    password.send_keys(Sheet.cell(row=i, column=8).value)                      #password
    loginbutton =driver.find_element_by_id(Sheet.cell(row=i, column=9).value)  #loginbutton
    loginbutton.click()                                                        #click login button


#code for category 4
def category_four(i):
    driver = webdriver.Chrome(options=chrome_options)
    driver.maximize_window()
    driver.get(Sheet.cell(row=i, column=3).value)                              #website 
    email =driver.find_element_by_name(Sheet.cell(row=i, column=4).value)      #email field by name
    email.send_keys(Sheet.cell(row=i, column=5).value)                         #email                                                     #click next button
    password =driver.find_element_by_name(Sheet.cell(row=i, column=7).value)   #password field by name
    password.send_keys(Sheet.cell(row=i, column=8).value)                      #password
    pyautogui.press('enter')                                                   #press ENTER

#code for category 5
def category_five(i):
    pyautogui.hotkey(Sheet.cell(row=i, column=10).value,Sheet.cell(row=i, column=11).value)

#code for category 6
def category_six(i):
    pyautogui.screenshot("myscreenshot.png")
    print("took screenshot")

#code for category 7
def category_seven(i):
    key1 = Sheet.cell(row=i, column=3).value
    print(key1,i)
    pyautogui.hotkey(key1,key1,key1,key1,key1,key1,key1,key1,key1,key1,key1,key1,key1,key1,key1,key1,key1,key1,key1,key1)
    

#code for category 8    
def category_eight(i):
    key1 = Sheet.cell(row=i, column=3).value
    key2 = Sheet.cell(row=i, column=4).value
    print(key1,key2)
    pyautogui.hotkey(key1,key2)

# wb= ex.load_workbook('D:/my project/laptop3.xlsx')      #loading excel workbook
wb= ex.load_workbook('D:/my project/major project/laptop.xlsx') 
Sheet = wb.get_sheet_by_name('Sheet1')  
    
    
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=True, max_num_hands=2, min_detection_confidence=0.5)
hands_videos = mp_hands.Hands(static_image_mode=False, max_num_hands=2, min_detection_confidence=0.5)
mp_drawing = mp.solutions.drawing_utils

def detectHandsLandmarks(image, hands, draw=True, display = True):
    # Create a copy of the input image to draw landmarks on.
    output_image = image.copy()
    
    # Convert the image from BGR into RGB format.
    imgRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Perform the Hands Landmarks Detection.
    results = hands.process(imgRGB)
    
    # Check if landmarks are found and are specified to be drawn.
    if results.multi_hand_landmarks and draw:
        
        # Iterate over the found hands.
        for hand_landmarks in results.multi_hand_landmarks:
            
            # Draw the hand landmarks on the copy of the input image.
            mp_drawing.draw_landmarks(image = output_image, landmark_list = hand_landmarks,
                                      connections = mp_hands.HAND_CONNECTIONS,
                                      landmark_drawing_spec=mp_drawing.DrawingSpec(color=(255,255,255), thickness=2, circle_radius=2),
                                      connection_drawing_spec=mp_drawing.DrawingSpec(color=(0,255,0), thickness=2, circle_radius=2))
    
    # Check if the original input image and the output image are specified to be displayed.
    if display:
        # Display the original input image and the output image.
        plt.figure(figsize=[15,15])
        plt.subplot(121);plt.imshow(image[:,:,::-1]);plt.title("Original Image");plt.axis('off');
        plt.subplot(122);plt.imshow(output_image[:,:,::-1]);plt.title("Output");plt.axis('off');
        
    # Otherwise
    else:
        
        # Return the output image and results of hands landmarks detection.
        return output_image, results 
    
    
def countFingers(image, results, draw=False, display=False):
    # Get the height and width of the input image.
    height, width, _ = image.shape
    
    # Create a copy of the input image to write the count of fingers on.
    output_image = image.copy()
    
    # Initialize a dictionary to store the count of fingers of both hands.
    count = {'RIGHT': 0, 'LEFT': 0}
    
    # Store the indexes of the tips landmarks of each finger of a hand in a list.
    fingers_tips_ids = [mp_hands.HandLandmark.INDEX_FINGER_TIP, mp_hands.HandLandmark.MIDDLE_FINGER_TIP,
                        mp_hands.HandLandmark.RING_FINGER_TIP, mp_hands.HandLandmark.PINKY_TIP]
    
    # Initialize a dictionary to store the status (i.e., True for open and False for close) of each finger of both hands.
    fingers_statuses = {'RIGHT_THUMB': False, 'RIGHT_INDEX': False, 'RIGHT_MIDDLE': False, 'RIGHT_RING': False,
                        'RIGHT_PINKY': False, 'LEFT_THUMB': False, 'LEFT_INDEX': False, 'LEFT_MIDDLE': False,
                        'LEFT_RING': False, 'LEFT_PINKY': False}
    
    
    # Iterate over the found hands in the image.
    for hand_index, hand_info in enumerate(results.multi_handedness):
        
        # Retrieve the label of the found hand.
        hand_label = hand_info.classification[0].label
        
        # Retrieve the landmarks of the found hand.
        hand_landmarks =  results.multi_hand_landmarks[hand_index]
        
        # Iterate over the indexes of the tips landmarks of each finger of the hand.
        for tip_index in fingers_tips_ids:
            
            # Retrieve the label (i.e., index, middle, etc.) of the finger on which we are iterating upon.
            finger_name = tip_index.name.split("_")[0]
            
            # Check if the finger is up by comparing the y-coordinates of the tip and pip landmarks.
            if (hand_landmarks.landmark[tip_index].y < hand_landmarks.landmark[tip_index - 2].y):
                
                # Update the status of the finger in the dictionary to true.
                fingers_statuses[hand_label.upper()+"_"+finger_name] = True
                
                # Increment the count of the fingers up of the hand by 1.
                count[hand_label.upper()] += 1
        
        # Retrieve the y-coordinates of the tip and mcp landmarks of the thumb of the hand.
        thumb_tip_x = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_TIP].x
        thumb_mcp_x = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_TIP - 2].x
        
        # Check if the thumb is up by comparing the hand label and the x-coordinates of the retrieved landmarks.
        if (hand_label=='Right' and (thumb_tip_x < thumb_mcp_x)) or (hand_label=='Left' and (thumb_tip_x > thumb_mcp_x)):
            
            # Update the status of the thumb in the dictionary to true.
            fingers_statuses[hand_label.upper()+"_THUMB"] = True
            
            # Increment the count of the fingers up of the hand by 1.
            count[hand_label.upper()] += 1
    
    code=[]
    
    for i in fingers_statuses:
        if (  fingers_statuses[i] == True ) :
            code.append(1)
        else:
            code.append(0)
            
    str1=""
    for i in code:
        str1+=str(i)
    print("Detected Gesture Code: ",str1)
    
    for i in range(1,Sheet.max_row+1):
        if Sheet.cell(row=i, column=1).value==str1:
            if (Sheet.cell(row=i, column=2).value)==1:               #category1: simple go to page
                category_one(i) #code for category 2
            elif (Sheet.cell(row=i, column=2).value)==2:             #category2: go to page - email - password - login
                category_two(i) #code for category 2
            elif (Sheet.cell(row=i, column=2).value)==3:             #category3: go to page - email - next - password - login
                category_three(i) #code for category 3
            elif (Sheet.cell(row=i, column=2).value)==4:             #category4: go to page - email(by name) - password(by name) - login
                category_four(i) #code for category 4
            elif (Sheet.cell(row=i, column=2).value)==5:             #category4: 2keys shortcut
                category_five(i) #code for category 5
            elif (Sheet.cell(row=i, column=2).value)==6:             #category4: 3 key shortcut
                category_six(i) #code for category 6
            elif (Sheet.cell(row=i, column=2).value)==7:             #category7: 3 key shortcut
                category_seven(i) #code for category 6 
            elif (Sheet.cell(row=i, column=2).value)==8:             #category8: 3 key shortcut
                category_eight(i) #code for category 6
            break
    else:
        print("string not found")
        
    # Check if the total count of the fingers of both hands are specified to be written on the output image.
    if draw:
        # Write the total count of the fingers of both hands on the output image.
        cv2.putText(output_image, " Total Fingers: ", (10, 25),cv2.FONT_HERSHEY_COMPLEX, 1, (20,255,155), 2)
        cv2.putText(output_image, str(sum(count.values())), (width//2-150,240), cv2.FONT_HERSHEY_SIMPLEX, 8.9, (20,255,155), 10, 10)

    # Check if the output image is specified to be displayed.
    if display:
        # Display the output image.
        plt.figure(figsize=[10,10])
        plt.imshow(output_image[:,:,::-1]);plt.title("Output Image");plt.axis('off');
        
    # Otherwise
    else:
        # Return the output image, the status of each finger and the count of the fingers up of both hands.
        return output_image, fingers_statuses, count
    
# Initialize the VideoCapture object to read from the webcam.
camera_video = cv2.VideoCapture(0)
camera_video.set(3,1280)
camera_video.set(4,960)
# Create named window for resizing purposes.
cv2.namedWindow('Fingers Counter', cv2.WINDOW_NORMAL)
# Iterate until the webcam is accessed successfully.
while camera_video.isOpened():
    time.sleep(2)
    ok, frame = camera_video.read()                                               # Read a frame   
    frame = cv2.flip(frame, 1)                                                    # Flip the frame horizontally for natural (selfie-view) visualization.
    frame, results = detectHandsLandmarks(frame, hands_videos, display=False)     # Perform Hands landmarks detection on the frame.
    
    # Check if the hands landmarks in the frame are detected.
    if results.multi_hand_landmarks:
        frame, fingers_statuses, count = countFingers(frame, results, display=False)        # Count the number of fingers up of each hand in the frame.
    cv2.imshow('Fingers Counter', frame)          # Display the frame.
    k = cv2.waitKey(1) & 0xFF          # Wait for 1ms. If a key is pressed, retreive the ASCII code of the key.
    if(k == 27):                # Check if 'ESC' is pressed and break the loop.
        break
```
